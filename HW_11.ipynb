Даны две функции: 
_f(x) = x^3 - 50x_ и _g(x) = -x^4 + 88x^2 - 241_

Требуется:

1. Найти координаты точек пересечения (Прим: в числовом виде, а не в символьном).
2. Построить графики функций в одной системе координат.
3. Построить график функции пересечения (из пункта 1)

```python
from sympy import *
from sympy.plotting import plot
import matplotlib
```

```python
x = Symbol('x')
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
```

1. КООРДИНАТЫ ТОЧЕК ПЕРЕСЕЧЕНИЯ

```python
res = solve(f - g)
```

```python
for elem in range(len(res)):
    coord_x = round(res[elem].n(), 2)
    coord_y = f.subs(x, coord_x)
    print(f'{elem + 1}-ая точка пересечения\n x: {coord_x}, y: {coord_y}\n')
```

2. ГРАФИКИ ФУНКЦИЙ В ОДНОЙ СИСТЕМЕ КООРДИНАТ

```python
task2 = plot(f, g)
```

3. ГРАФИК ФУНКЦИИ ПЕРЕСЕЧЕНИЯ

```python
task3 = plot(f - g)
```
